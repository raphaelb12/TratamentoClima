```python
import pandas as pd
import statistics  as sts
```

```python
#importa dados
dataset = pd.read_csv("tempo.csv", sep=";")
#visulizar
#dataset.head()
dataset
```

```python
#dataset com dados de dia de 'Aparencia = Chuva' para tratar o NaN em Umidade
datasetchuva = dataset.loc[dataset['Aparencia'] == 'chuva']
datasetchuva
```

```python
#mediana da umidade em dias de chuva
mediana = sts.median(datasetchuva['Umidade'])
mediana
```

```python
#substituindo o NaN pela mediana
dataset['Umidade'].fillna(mediana, inplace = True)
dataset
```

```python
#agora tratando o outlier
dataset.loc[(dataset['Umidade'] < 0) | (dataset['Umidade'] > 100)]
```

```python
#atribuindo qualquer outlier ao valor da mediana
dataset.loc[(dataset['Umidade'] < 0) | (dataset['Umidade'] > 100), 'Umidade'] = mediana
```

```python
#verificando que não existem mais outliers
dataset.loc[(dataset['Umidade'] < 0) | (dataset['Umidade'] > 100)]
```

```python
dataset
```

```python
#checando a quantidade de verdadeiros e falsos em 'Vento'
numvf = dataset.groupby(['Vento']).size()
numvf
```

```python
#checando elementos null em 'Vento'
vento_null = dataset.loc[dataset['Vento'].isnull()]
vento_null
```

```python
#substituindo o NaN em 'Vento' pela moda
dataset['Vento'].fillna('FALSO', inplace = True)
dataset
```

```python
#verificando os outliers em 'Temperatura'
dataset.loc[(dataset['Temperatura'] > 130) | (dataset['Temperatura'] < -130)]
```

```python
#verificando todos os dados em dias de sol
datasetsol = dataset.loc[dataset['Aparencia'] == 'sol']
datasetsol
```

```python
datasetsol.describe()
```

```python
#guardando a mediana numa variavel
medianasol = sts.median(datasetsol['Temperatura'])
medianasol
```

```python
#substituindo o outlier pela mediana em um dia de sol
dataset.loc[(dataset['Temperatura'] < -130) | (dataset['Temperatura'] > 130), 'Temperatura'] = medianasol
```
